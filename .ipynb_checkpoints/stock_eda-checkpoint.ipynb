```python
## Getting data from stockrow instead - has quarterly data which is better - more data points
import urllib.request

url_fq = "https://stockrow.com/api/companies/AAPL/financials.xlsx?dimension=MRQ&section=Income%20Statement&sort=desc"
response = urllib.request.urlopen(url_fq)
financials_q = pd.read_excel(response) 
financials_q = financials_q.T
url_ft = "https://stockrow.com/api/companies/AAPL/financials.xlsx?dimension=MRT&section=Income%20Statement&sort=desc"
response = urllib.request.urlopen(url_ft)
financials_t = pd.read_excel(response) 
financials_t = financials_t.T
url_mq = "https://stockrow.com/api/companies/AAPL/financials.xlsx?dimension=MRQ&section=Metrics&sort=desc"
response = urllib.request.urlopen(url_mq)
metrics_q = pd.read_excel(response)
metrics_q = metrics_q.T
```

```python
url_mt = "https://stockrow.com/api/companies/AAPL/financials.xlsx?dimension=MRT&section=Metrics&sort=desc"
response = urllib.request.urlopen(url_mt)
metrics_t = pd.read_excel(response)
metrics_t = metrics_t.T
```
